#### Questão 3

Em seguida montar um gráfico com a melhor visualização para os dados com a biblioteca Plotly Express, demonstrando relação entre as variáveis.

````python
obs3 = np.random.randint(1000, 2001, 500)
obs4 = np.random.randint(5000, 10001, 500)
````
