```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

#### Questão 1

Qual o valor médio da gorjeta (*tip*) de acordo com o dia da semana (*day*)?

```python
sns.barplot(data=tips, x='day', y='tip', estimator='mean', errorbar=None)

# Estética
plt.title("Average tip amount per day")
plt.xlabel('Day')
plt.ylabel('Average tip amount')
plt.grid()

# show
plt.show()
```

#### Questão 2

Qual a soma das gorjetas (*tip*) por turno do dia (*time*), observando os fumantes e não fumantes (*smoker*)?

```python
tips.head()
```

```python
#sns.barplot(data=tips, x='time', y='tip', hue='smoker', estimator='sum', errorbar=None)
ax = sns.barplot(data=tips, 
                    x='time', 
                    y='tip', 
                    hue='smoker', 
                    estimator='sum', 
                    errorbar=None)

for i in ax.containers:
    ax.bar_label(i,)

plt.title('Sum of tips per time between smokers and non-smokers')
plt.xlabel('Time')
plt.ylabel('Sum of tips')
plt.legend(title='Smoker', loc='upper left', bbox_to_anchor=(1, 1))

plt.show()
```

#### Questão 3

Verique a distribuição de gorjetas (*tip*) por dia (*day*) e turno (*time*). Há algum dia e turno com maior presença de outliers?

```python
tips.head()
```

```python
sns.boxplot(data=tips, y='tip', x='day', hue='time')
plt.title('Distribution of tips throughout the day and time')
plt.xlabel('Day')
plt.ylabel('Tips')
plt.legend(title='Time', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
#Aos sábados no jantar há maior presença de outliers
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')

# Estética
plt.title('Relation between Total bill and Tips')
plt.xlabel('Total bill')
plt.ylabel('Tips')
plt.grid()
plt.show()
```
